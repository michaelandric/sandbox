```python
import numpy as np
from sklearn import mixture

import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline
```

```python
n_samples = 300
```

```python
np.random.seed(0)
```

```python
shifted_gaussian = np.random.randn(n_samples, 2) + np.array([20, 20])
```

```python
plt.scatter(shifted_gaussian[:, 0], shifted_gaussian[:, 1])
```

```python
# centered, stretched gaussian
C = np.array([[0., -0.7], [3.5, 7.]])
stretched_gaussian = np.dot(np.random.randn(n_samples, 2), C)
```

```python
plt.scatter(stretched_gaussian[:, 0], stretched_gaussian[:, 1])
```

```python
X_train = np.vstack([shifted_gaussian, stretched_gaussian])
```

```python
# two components to the gaussian mixture model
clf = mixture.GaussianMixture(n_components=2, covariance_type='full')
clf.fit(X_train)
```

```python
# display predicted scores as a contour plot
x = np.linspace(-20., 30.)
y = np.linspace(-20., 40.)
X, Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T
Z = -clf.score_samples(XX)
Z = Z.reshape(X.shape)
```

```python
plt.figure(figsize=(8, 5))
CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0), levels=np.logspace(0, 3, 10))
CB = plt.colorbar(CS, shrink=0.8, extend='both')

plt.scatter(X_train[:, 0], X_train[:, 1], .5)

plt.title('Negative log-likelihood predicted by a GMM')
plt.axis('tight')
plt.show()
```

## py-earth 

```python
from pyearth import Earth
from sklearn.ensemble import RandomForestRegressor
```

```python
np.random.seed(2)
```

```python
m = 10000
n = 10

X = np.random.uniform(size=(m, n))
```

```python
y = (10 * np.sin(np.pi * X[:, 0] * X[:, 1]) +
     20 * (X[:, 2] - 0.5) ** 2 +
     10 * X[:, 3] +
     5 * X[:, 4] + np.random.uniform(size=m))
```

### Fit an Earth model

```python
criteria = ('rss', 'gcv', 'nb_subsets')
model = Earth(max_degree=3,
              max_terms=10,
              minspan_alpha=.5,
              feature_importance_type=criteria,
              verbose=True)
```

```python
model.fit(X, y)
```

```python
rf = RandomForestRegressor()
rf.fit(X, y)
```

```python
print(model.trace())
```

```python
print(model.summary())
```

```python
print(model.summary_feature_importances(sort_by='gcv'))
```

#### plot the feature importances

```python
importances = model.feature_importances_
```

```python
importances['random_forest'] = rf.feature_importances_
```

```python
criteria = criteria + ('random_forest',)
```

```python
idx = 1
```

```python
fig = plt.figure(figsize=(20, 10))
labels = ['$x_{}$'.format(i) for i in range(n)]
for crit in criteria:
    plt.subplot(2, 2, idx)
    plt.bar(np.arange(len(labels)),
            importances[crit],
            align='center',
            color='red')
    plt.xticks(np.arange(len(labels)), labels)
    plt.title(crit)
    plt.ylabel('importances')
    idx += 1
title = '$x_0,...x_9 \sim \mathcal{N}(0, 1)$\n$y= 10sin(\pi x_{0}x_{1}) + 20(x_2 - 0.5)^2 + 10x_3 + 5x_4 + Unif(0, 1)$'
fig.suptitle(title, fontsize="x-large")
plt.show()
```

## one class svm

```python
from sklearn.datasets import load_boston
```

```python
from sklearn.svm import OneClassSVM
```

```python
classifiers = {"OCSVM": OneClassSVM(nu=0.261, gamma=0.05),
               "OCSVM_2": OneClassSVM(nu=0.15, gamma=0.05)}
colors = ['magenta', 'cyan']
legend = {}
```

```python
# Get data
X1 = load_boston()['data'][:, [8, 10]]  # two clusters
X2 = load_boston()['data'][:, [5, 12]]  # "banana"-shaped
```

```python
xx1, yy1 = np.meshgrid(np.linspace(-8, 28, 500), np.linspace(3, 40, 500))
xx2, yy2 = np.meshgrid(np.linspace(3, 10, 500), np.linspace(-5, 45, 500))
```

```python
import matplotlib.font_manager

for i, (clf_name, clf) in enumerate(classifiers.items()):
    plt.figure(1)
    clf.fit(X2)
    Z2 = clf.decision_function(np.c_[xx2.ravel(), yy2.ravel()])
    Z2 = Z2.reshape(xx2.shape)
    
    legend[clf_name] = plt.contour(xx2, yy2, Z2, levels=[0],
                                   linewidths=2, colors=colors[i])

legend_values_list = list(legend.values())
legend_keys_list = list(legend.keys())

plt.figure(1)  # "banana" shape
plt.title("Outlier detection on boston housing data set")
plt.scatter(X2[:, 0], X2[:, 1], color='black', alpha=0.25)
#plt.contour(xx2, yy2, Z2, levels=[0], linewidths=2, colors=colors[i])
plt.xlim((xx2.min(), xx2.max()))
plt.ylim((yy2.min(), yy2.max()))

plt.legend((legend_values_list[0].collections[0],
            legend_values_list[1].collections[0]),
           (legend_keys_list[0], legend_keys_list[1]),
           frameon=False,
           prop=matplotlib.font_manager.FontProperties(size=12))

plt.ylabel("% lower status of the population")
plt.xlabel("average number of rooms per dwelling")

plt.show()
```
